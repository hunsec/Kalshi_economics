```python
import requests

# Replace with your actual FRED API key
api_key = "d3d05e602a5f8ce27ca19e908581410d"

# FRED series ID for the effective federal funds rate
series_id = "FEDFUNDS"

# FRED API endpoint for series observations
url = f"https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={api_key}&observation_start=2018-01-01&file_type=json"

# Fetch the data
response = requests.get(url)
data = response.json()

# Extract and display the daily federal funds rate since 2018
daily_fed_funds = [
    (obs["date"], float(obs["value"]))
    for obs in data.get("observations", [])
    if obs.get("value") and obs.get("value") != "."
]

# Print the first 5 and last 5 entries as a sample
print("Sample entries (first 5):")
for date, rate in daily_fed_funds[:5]:
    print(f"{date}: {rate}%")

print("\nSample entries (last 5):")
for date, rate in daily_fed_funds[-5:]:
    print(f"{date}: {rate}%")
```

```python
import pandas as pd


# Convert the list of tuples to a pandas DataFrame
df = pd.DataFrame(daily_fed_funds, columns=["Date", "Federal Funds Rate (%)"])

# Save the DataFrame to an Excel file
excel_file = "C:\\Users\\lakch\\Documents\\federal_funds_rate_2018_present.xlsx"
df.to_excel(excel_file, index=False)

print(f"Data saved to {excel_file}")
```
