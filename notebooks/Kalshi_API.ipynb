```python
import requests

# Get series information for KXECONSTATCORECPIYOY
url = "https://api.elections.kalshi.com/trade-api/v2/series/KXCPI"
response = requests.get(url)
series_data = response.json()

# Check if 'series' key exists before accessing its values
if 'series' in series_data:
    print(f"Series Title: {series_data['series']['title']}")
    print(f"Frequency: {series_data['series']['frequency']}")
    print(f"Category: {series_data['series']['category']}")
else:
    print(f"Error fetching series data: {series_data.get('error', 'Unknown error')}")
print(series_data)
```

```python
# Get all markets for the KXECONSTATCORECPIYOY series (removing the status filter)
markets_url = f"https://api.elections.kalshi.com/trade-api/v2/markets?series_ticker=KXCPI"
markets_response = requests.get(markets_url)
markets_data = markets_response.json()

print(f"\nActive markets in KXECONSTATCORECPIYOY series:")
for market in markets_data['markets']:
    print(f"- {market['ticker']}: {market['title']}")
    print(f"  Event: {market['event_ticker']}")
    # Use .get() to safely access 'yes_price' and 'volume', providing 'N/A' if not present
    yes_price_display = f"{market.get('yes_price', 'N/A')}¢" if market.get('yes_price') is not None else 'N/A'
    volume_display = market.get('volume', 'N/A')
    print(f"  Yes Price: {yes_price_display} | Volume: {volume_display}")
    print()

# Get details for a specific event if you have its ticker
if markets_data['markets']:
    # Let's get details for the first market's event
    event_ticker = markets_data['markets'][0]['event_ticker']
    event_url = f"https://api.elections.kalshi.com/trade-api/v2/events/{event_ticker}"
    event_response = requests.get(event_url)
    event_data = event_response.json()

    print(f"Event Details:")
    print(f"Title: {event_data['event']['title']}")
    print(f"Category: {event_data['event']['category']}")
```

```python
import requests

# Get all active markets from Kalshi API for the KXCPI series
# This code is added to ensure 'markets_data' is defined when this cell is executed.
markets_url = "https://api.elections.kalshi.com/trade-api/v2/markets?series_ticker=KXCPI"
markets_response = requests.get(markets_url)
markets_data = markets_response.json()

# Get orderbook for all markets in the KXCPI series

if not markets_data['markets']:
    print("No markets found for the KXCPI series. Please ensure the series ticker is correct and markets exist.")
else:
    for market in markets_data['markets']:
        market_ticker = market['ticker']
        market_title = market['title'] # Get the market title

        # Safely get the expected CPI core percentage if available
        # This field might not exist in all market types, so using .get() is robust.
        expected_cpi_core = market.get('custom_strike', {}).get('Value', 'N/A')

        orderbook_url = f"https://api.elections.kalshi.com/trade-api/v2/markets/{market_ticker}/orderbook"

        orderbook_response = requests.get(orderbook_url)
        orderbook_data = orderbook_response.json()

        print(f"\n--- Orderbook for {market_ticker} ---")
        print(f"Market Title: {market_title}")
        print(f"Expected CPI Core: {expected_cpi_core}%") # Print the expected CPI core

        print("YES BIDS:")
        if 'yes_dollars' in orderbook_data['orderbook_fp'] and orderbook_data['orderbook_fp']['yes_dollars']:
            for bid in orderbook_data['orderbook_fp']['yes_dollars'][:5]:  # Show top 5
                print(f"  Price: {bid[0]}¢, Quantity: {bid[1]}")
        else:
            print("  No YES bids available.")

        print("\nNO BIDS:")
        if 'no_dollars' in orderbook_data['orderbook_fp'] and orderbook_data['orderbook_fp']['no_dollars']:
            for bid in orderbook_data['orderbook_fp']['no_dollars'][:5]:  # Show top 5
                print(f"  Price: {bid[0]}¢, Quantity: {bid[1]}")
        else:
            print("  No NO bids available.")
        print("-------------------------------------")
```

```python
import requests
from datetime import datetime, UTC
# Get historical markets for the KXCPI series with 'settled' status
# We'll use a max_settle_time of today to get all markets settled up to now.
current_time_utc = datetime.now(UTC).isoformat() # Using timezone-aware object for UTC

historical_markets_url_base = "https://api.elections.kalshi.com/trade-api/v2/historical/markets/"
historical_markets_url = f"{historical_markets_url_base}KXCPI"
historical_markets_response = requests.get(historical_markets_url)

# Print the raw response text to debug the JSONDecodeError
print("Raw API Response:")
print(historical_markets_response.text)

try:
    historical_markets_data = historical_markets_response.json()

    print(f"\nHistorical Settled Markets in KXCPI series:")

    if 'markets' in historical_markets_data and historical_markets_data['markets']:
        print("List of Inflation/CPI-related Tickers and their details:")
        for market in historical_markets_data['markets']:
            print(f"  - Ticker: {market['ticker']}")
            print(f"    Title: {market['title']}")
            print(f"    Event: {market['event_ticker']}")
            print(f"    Close Time: {market.get('close_time', 'N/A')}")
            print(f"    Settlement Price: {market.get('settlement_price', 'N/A')}¢")
            print(f"    Settlement Value: {market.get('settlement_value', 'N/A')}")
            print()
    else:
        print("No historical settled markets found for the KXCPI series.")
except requests.exceptions.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
    print("Could not parse the API response as JSON. The raw response is printed above.")
```

```python
import requests

# 1. Get all active markets for the KXCPI series to obtain individual market tickers
markets_url = "https://api.elections.kalshi.com/trade-api/v2/markets?series_ticker=KXCPI"
markets_response = requests.get(markets_url)
markets_data = markets_response.json()

print("\n--- Actieve KXCPI markten ophalen voor individuele historische aanroepen ---")
if 'markets' in markets_data and markets_data['markets']:
    kxcpi_market_tickers = [market['ticker'] for market in markets_data['markets']]
    print(f"{len(kxcpi_market_tickers)} actieve KXCPI markten gevonden. Nu individuele historische data ophalen...")

    historical_markets_url_base = "https://api.elections.kalshi.com/trade-api/v2/historical/markets/"

    print("\n--- Historische data voor individuele KXCPI markten --- ")
    for ticker in kxcpi_market_tickers:
        individual_historical_url = f"{historical_markets_url_base}{ticker}"
        individual_historical_response = requests.get(individual_historical_url)

        # Check if the request was successful and if there's content to parse
        if individual_historical_response.status_code == 200 and individual_historical_response.text:
            try:
                individual_historical_data = individual_historical_response.json()
                if 'market' in individual_historical_data:
                    market_info = individual_historical_data['market']
                    print(f"  - Ticker: {market_info.get('ticker', 'N/A')}")
                    print(f"    Title: {market_info.get('title', 'N/A')}")
                    # You could add more fields here if needed, e.g., print(f"    Status: {market_info.get('status', 'N/A')}")
                else:
                    print(f"  Geen market details gevonden voor ticker {ticker} in historische response.")
            except requests.exceptions.JSONDecodeError:
                print(f"  Fout bij decoderen van JSON voor historische data van {ticker}. Ruwe respons: {individual_historical_response.text[:200]}...")
        elif individual_historical_response.status_code == 404:
            print(f"  Historische data voor ticker {ticker} niet gevonden (404 Not Found).")
        else:
            print(f"  Fout bij het ophalen van historische data voor ticker {ticker}. Statuscode: {individual_historical_response.status_code}")
            print(f"  Ruwe respons: {individual_historical_response.text[:200]}...")

else:
    print("Geen actieve markten gevonden voor de KXCPI-serie.")
```
